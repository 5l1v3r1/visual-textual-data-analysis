```python
import pandas as pd
```

```python
pid = df['PID']
print('len of PID', len(pid))
p_exist = df['p_exist']
b_exist = df['b_exist']
q1 = '/Users/mayanksaxena/Desktop/gesture-analysis/fixation/' 
q2 = '/Output/' 
q3 = '.xls'
two = '2_'
three = '3_'
one = '1_'

b_top_left_x = 233
b_top_left_y = 74
b_bottom_right_x = 830 
b_bottom_right_y = 414

b_image_top_x = 620
b_image_top_y = 70
b_image_bottom_x = 830
b_image_bottom_y = 420

b_speaker_top_x = 84
b_speaker_top_y = 232
b_speaker_bottom_x = 668
b_speaker_bottom_y = 945

p_top_left_x = 530
p_top_left_y = 190
p_bottom_right_x = 1680 
p_bottom_right_y = 900

p_image_top_x = 1130
p_image_top_y = 240
p_image_bottom_x = 1657
p_image_bottom_y = 860

p_speaker_top_x = 120                                                                                                                                                                                                                                                                                                                                      
p_speaker_top_y = 300
p_speaker_bottom_x = 600
p_speaker_bottom_y = 920
```

```python
k = 0

for i in range(0, len(pid)):
    s2 = q1 + two + str(pid[i]) + q2 + two + str(pid[i]) + q3
    s1 = q1 + three + str(pid[i]) + q2 + three + str(pid[i]) + q3
    print('s1', s1)
    print('s2', s2)
    
    if (int(p_exist[i]) == 1):
        print('here here')
        k += 1
        df = pd.read_csv(s2, sep='\t')
        
        #print('s1', s1)
        df.head()
        
        x = df['CURRENT_FIX_X']
        y = df['CURRENT_FIX_Y']
        frame_start = df['VIDEO_FRAME_INDEX_START']
        frame_end = df['VIDEO_FRAME_INDEX_END']
        
        
        in_slide = []
        in_image = []
        in_text = []
        in_speaker = []
        
        print('length of x', len(x))
        print('length of y', len(y))

        for j in range(len(x)):
            x_coord = int(x[j])
            y_coord = int(y[j])
    
            if ((b_top_left_x <= x_coord <= b_bottom_right_x) and (b_top_left_y <= y_coord <= b_bottom_right_y)):
                # Point is within the slide
                in_slide.append(1)
            else:
                in_slide.append(0)
            
        for j in range(len(x)):
            x_coord = int(x[j])
            y_coord = int(y[j])
            
            if ((b_image_top_x <= x_coord <= b_image_bottom_x) and (b_image_top_y <= y_coord <= b_image_bottom_y)):
                # Point is within the slide and on image
                in_image.append(1)
                in_text.append(0)
            else:
                # Point is within the slide and on text
                if (in_slide[j] == 1):
                    in_text.append(1)
                    in_image.append(0)
                # Point is within the slide and neither on image nor text
                else:
                    in_text.append(0)
                    in_image.append(0)
                    
            if ((b_speaker_top_x <= x_coord <= b_speaker_bottom_x) and (b_speaker_top_y <= y_coord <= b_speaker_bottom_y)):
                # Point is not on the speaker
                in_speaker.append(0)
            else:
                # Point is on the speaker
                in_speaker.append(1)
        
        df_new = pd.DataFrame()
        df_new['x_coord'] = x
        df_new['y_coord'] = y    
        df_new['in_slide'] = in_slide
        df_new['in_text'] = in_text
        df_new['in_image'] = in_image
        df_new['in_speaker'] = in_speaker
        df_new['starting_frame'] = frame_start
        df_new['ending_frame'] = frame_end
       
        
        print(df_new.head())
        
        #p = str(pid[i])
        try:
            print('i', i)
            print('k', k)
            file_name = 'p_data/' + '2_' + str(pid[i]) + '_data.csv' 
            print('Filename', file_name)
            df_new.to_csv(file_name, encoding='utf-8', index=False)
            print('count of in_slide', len(in_slide))
            print('count of in_text', len(in_text))
            print('count of in_image', len(in_image))
            print('count of in_speaker', len(in_speaker))
        except Exception as e:
            print('i', i)
            print('exception', str(e))
            print('in exception')
            
        #break
```
