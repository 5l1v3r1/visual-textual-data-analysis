```python
import os
from fnmatch import fnmatch

root = '/Users/mayanksaxena/Desktop/gesture-analysis/fixation/'
pattern = "*.xls"
i = 0

list_of_files = []

for path, subdirs, files in os.walk(root):
    for name in files:
        if fnmatch(name, pattern):
            s = os.path.join(path, name)
            #print(os.path.join(path, name))
            print(s)
            i +=1
            list_of_files.append(s)
print('i', i)
```
