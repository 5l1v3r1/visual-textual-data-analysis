```python
#df = pd.read_csv('test_data.csv')
df = pd.read_csv('3r1-test.csv')
df.head(5)
x = df['CURRENT_FIX_X']
y = df['CURRENT_FIX_Y']
len(x)
```

```python
in_slide = []

x = df['CURRENT_FIX_X']
y = df['CURRENT_FIX_Y']

top_left_x = 233
top_left_y = 74
bottom_right_x = 830 
bottom_right_y = 414
for i in range(len(x)):
    x_coord = int(x[i])
    y_coord = int(y[i])
    
    if ((top_left_x <= x_coord <= bottom_right_x) and (top_left_y <= y_coord <= bottom_right_y)):
        # Point is within the slide
        in_slide.append(1)
    else:
        in_slide.append(0)

df['in_slide'] = in_slide
df.head()
```

```python
in_image = []
in_text = []
in_slide = df['in_slide']

image_top_x = 620
image_top_y = 70
image_bottom_x = 830
image_bottom_y = 420

for i in range(len(x)):
    x_coord = int(x[i])
    y_coord = int(y[i])

    if ((image_top_x <= x_coord <= image_bottom_x) and (image_top_y <= y_coord <= image_bottom_y)):
    # Point is within the slide
        in_image.append(1)
        in_text.append(0)
    else:
        if (in_slide[i] == 1):
            in_text.append(1)
            in_image.append(0)
        else:
            in_text.append(0)
            in_image.append(0)

df['in_image'] = in_image
df['in_text'] = in_text

df.head()
```
